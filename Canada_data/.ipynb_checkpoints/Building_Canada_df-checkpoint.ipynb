```python
# MAkes a list and prints every station that has no null data and That has existed in 1917 and 2017
station_list = []
for station in list(df1917['Stn_Name']):
    if station in list(df2017['Stn_Name']):
        station_list.append(station)
        print (station)
   
```

```python
# Picked 5 station that are in diffrent parts of Canada.
# These are the stations we are going to build our dataframe out of.
stations =['OTTAWA CDA', 'QUATSINO', 'INDIAN BAY', 'KASLO', 'ATLIN']
```

```python
# Makes a function to get thge station data from a csv, and turns it into a dataframe. 
def process_data(file):
    stations=['OTTAWA CDA', 'QUATSINO', 'INDIAN BAY', 'KASLO','ATLIN']
    df1 = pd.read_csv(file, skiprows=30)
    df1= df1[df1['Stn_Name'].isin(stations)]
    df1['Year'] = int(file.split('_')[-2])
    df1['Month'] = int(file.split('_')[-1].split('.')[0])
    return df1
```

```python
# Grabs the data from each station in staion list, from each csv. 
canada_df_list = []
path = './Datasets/DataWorld/Canada1917-2017/'
for file in os.listdir(path):
    canada_df_list.append(process_data(path + file))
```

```python
# Concatieates all the dataframes in the canada_df_list inro one df
canada_df_raw = pd.concat(canada_df_list, ignore_index=True)
```

```python
# Makes a dictionary showing what measure of precipitaion 'P%N' uses by month and station 
P_dict = {}
for station in canada_df['Stn_Name'].unique():
    P_dict[station] = {}
    for month in range(1,13):
        avg_P = canada_df[(canada_df['Month'] == month) & (canada_df['Stn_Name']== station)]['P'].mean()
        avg_pctN = canada_df[(canada_df['Month'] ==month) & (canada_df['Stn_Name']== station)]['P%N'].mean()
        normal_P = avg_P/avg_pctN 
        P_dict[station][month] = normal_P
    print(f'{station} {month}', normal_P) 
```

```python
# Shows all rows with missing 'P%N' values
canada_df[canada_df['P%N'] != canada_df['P%N']]
```

```python
# turns the indexs of the rows with missing 'P%N' values into a list
missing_pctN_idx = list(canada_df[canada_df['P%N'] != canada_df['P%N']].index)
```

```python
# The for loop replaces the missing 'P%N' vales with 
# values based on station, month, and precipitation for that row
for idx in missing_pctN_idx:
        station = canada_df.at[idx,'Stn_Name']
        month = canada_df.at[idx,'Month']
            
        target_p = P_dict[station][month]
        precipitation = canada_df.at[idx,'P']

        pctN = precipitation / target_p 
        canada_df.at[idx,'P%N'] = pctN
```

```python
canada_df[canada_df['S%N'] != canada_df['S%N']]
```

```python
# turns the indexs of the rows with missing 'S%N' values into a list
missing_pctS_idx = list(canada_df[canada_df['S%N'] != canada_df['S%N']].index)
```

```python
# Makes a dictionary showing what measure of precipitaion 'S%N' uses by month and station 
S_dict = {}
for station in canada_df['Stn_Name'].unique():
    S_dict[station] = {}
    for month in range(1,13):
        avg_S = canada_df[(canada_df['Month'] == month) & (canada_df['Stn_Name']== station)]['S'].mean()
        avg_pctS = canada_df[(canada_df['Month'] ==month) & (canada_df['Stn_Name']== station)]['S'].mean()
        
        if avg_pctS != 0:
            normal_S = avg_S/avg_pctS 
            S_dict[station][month] = normal_S
        elif avg_pctS == 0:
            S_dict[station][month] = 0
    print(f'{station} {month}', normal_S) 
```

```python
# The for loop replaces the missing 'S%N' vales with 
# values based on station, month, and precipitation for that row
for idx in missing_pctS_idx:
        station = canada_df.at[idx,'Stn_Name']
        month = canada_df.at[idx,'Month']
            
        target_S = S_dict[station][month]
        snowfall = canada_df.at[idx,'S']

        pctS = snowfall / target_S
        canada_df.at[idx,'S%N'] = pctS
```
